#### Ordinal encoding using sklearn and map_function

```python
df2 = pd.DataFrame({"Size":["s","x","xxl","m","s","s","x","m","xxl"]})
```

```python
Ord_data = [["x","s","m","xxl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=Ord_data)
oe.fit(df2[["Size"]])
```

```python
df2["Size_en"] = oe.transform(df2[["Size"]])
df2
```

```python
Ord_data = {"s":1,"m":3,"xxl":4,"x":5}
```

```python
df2["Size_en_map"] = df2["Size"].map(Ord_data)
```

```python
df2
```
