```python
df = pd.read_excel("groceries - groceries.xlsx")
df.head()
```

```python
 df.shape
```

```python
market = []
for i in range(0,df.shape[0]):
    cus = []
    for j in df.columns:
        if type(df[j][i])==str:
            cus.append(df[j][i])
    market.append(cus)
```

```python
l = []
for i in market:
    for j in i:
        l.append(j)
```

```python
import collections
```

```python
p = collections.Counter(l)
```
