```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_excel("placement1.xlsx")
df.head()
```

```python
sns.scatterplot(x="cgpa",y="resume_score",data=df,hue="placed")
plt.show()
```

```python
x = df.iloc[:,:-1]
y = df["placed"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

#### Hyper petameters(linear,poly,rbf,sigmoid)

```python
from sklearn.svm import SVC
sv = SVC(kernel="linear")
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)*100
```

```python
sv.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=sv)
plt.show()
```
