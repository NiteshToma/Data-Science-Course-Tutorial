```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("Social_Network_Ads.xlsx")
df.head(3)
```

```python
x = df.iloc[:,:-1]
y = df["Purchased"]
```

```python
df["Purchased"].value_counts()
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
ru = RandomUnderSampler()
ru_x,ru_y = ru.fit_resample(x,y)
```

```python
ru_y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(ru_x,ru_y,random_state=42,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[47,25000]])
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ro = RandomOverSampler()
ro_x,ro_y = ro.fit_resample(x,y)
```

```python
ro_y.value_counts()
```

```python
x_train1,x_test1,y_train1,y_test1 = train_test_split(ro_x,ro_y,random_state=42,test_size=0.2)
```

```python
lr1 = LogisticRegression()
lr1.fit(x_train1,y_train1)
```

```python
lr1.score(x_test1,y_test1)*100
```

```python
lr1.predict([[19,19000]])
```
